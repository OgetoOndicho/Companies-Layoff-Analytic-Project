```python
## Convert the Layoff_date column from OBJECT to DATE
employee_data['Layoff_date'] = pd.to_datetime(employee_data['Layoff_date'])
employee_data['Layoff_date']
                                              
```

```python
## PART B: ANSWERING BUSINESS QUESTIONS
## Q1. What are the changes of layoffs over time? Are there any specific periods with spikes?
### Lets extract the YEAR Column from the data
employee_data['Year'] = employee_data['Layoff_date'].dt.strftime('%Y')
employee_data
```

```python
## Having derived the YEAR From the data, we can group employees laid off by the year
yearly_layoffs = employee_data.groupby('Year')['total_laid_off'].sum().reset_index()
yearly_layoffs
```

```python
## TO Group By Year with the most employees laid off
yearly_layoffs.sort_values( by = 'total_laid_off', ascending = False)
```

```python
## Visualize the trend of Employees lay off at the company
plt.figure()
plt.xlabel('Year')
plt.ylabel('total_laid_off')
plt.plot(yearly_layoffs['Year'], yearly_layoffs['total_laid_off'])
plt.title('Layoffs Over Time')
plt.show()
```

```python
## Q2. Identify the industries with the Highest Number of Employee layoffs
company_layoffs = employee_data.groupby(['company', 'industry', 'location'])['total_laid_off'].sum().reset_index()
company_layoffs
```

```python
## Write code to identify the top company with the highest employees total lay offs
top_company = company_layoffs[company_layoffs['total_laid_off'] == company_layoffs['total_laid_off'].max()]
print(top_company)
```

```python
## Q3. Which countries and locations are employees highly affected with lay_offs
country_info = employee_data.groupby(['country', 'location'])['total_laid_off'].sum().reset_index()
country_info
```

```python
## Let us sort the data in order with Countries with most Layoffs at the Top
country_info.sort_values(by ='total_laid_off', ascending = False).head(5)
```

```python
## Q4. Is there relationship between the  Rate of high Employees lay offs and the percentage of employees laid off? 
### Use a scatter plot first to visualize if there is any possibility
plt.figure()
plt.xlabel('total_laid_off')
plt.ylabel('percentage_laid_off')
plt.title('Employees: Total Lay offs VS Percentage Lay offs')
plt.scatter(employee_data['total_laid_off'], employee_data['percentage_laid_off'])
plt.show()

```

```python
## Further dissect if there is any correlation between the  Rate of high Employees lay offs and the percentage of employees laid off?
layoff_correlation = employee_data['total_laid_off'].corr(employee_data['percentage_laid_off'])
print(layoff_correlation)
```

```python
## Q6. Compare by Industry to see how Big Companies and Small Companies compete
comparison_df = employee_data.groupby('industry').agg({
    'total_laid_off': 'sum',
    'percentage_laid_off': 'mean'
}).reset_index()

print(comparison_df)
```
